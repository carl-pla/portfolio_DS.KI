### Weinqualität vs. Alkoholgehalt

```python
# Benötigte Bibliotheken importieren
#import tensorflow as tf 
import pandas as pd
import matplotlib.pyplot as plt 

# CSV-Datei für Rotwein einlesen
df = pd.read_csv('/Users/carlplacek/Desktop/Uni/DataScience&KI/wine+quality/winequality-red.csv', sep=';')

""" 
# Überblick
print(df.head())   # erste Zeilen anzeigen
print(df.info())   # Spalten, Datentypen, fehlende Werte
"""

# Daten aus zwei Spalten extrahieren
x = df["alcohol"]   # unabhängige Variable (x-Achse)
y = df["quality"]   # abhängige Variable (y-Achse)

# Streudiagramm erstellen
plt.scatter(x, y, color='blue', marker='o', s=50)

# Durchschnitt berechnen
durchschnitt_alc = df["alcohol"].mean()
durchschnitt_qu = df["quality"].mean()

# Runden auf 2 Nachkommastellen
durchschnitt_gerundet_alc = round(durchschnitt_alc, 2)
durchschnitt_gerundet_qu = round(durchschnitt_qu, 2)

# Durchschnittswerte ausgeben als Ausgabe Befehl im Output
print("Durchschnitt Alkohlgehalt:", (durchschnitt_gerundet_alc), "%")
print("Durchschnitt Qualität:", (durchschnitt_gerundet_qu), "von 10")


# Diagramm mit pyplot visualisieren
plt.xlabel('Alcohol (%)')
plt.ylabel('Quality (0-10)')
plt.title('Alcohol vs Quality in Red Wine')
plt.grid(True)
plt.show()
```
